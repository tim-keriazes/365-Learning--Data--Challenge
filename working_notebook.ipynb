```python
# import necessary libraries
import pandas as pd
import os
import glob
  
  
# use glob to get all the csv files 
# in the folder
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
  
  
# loop over the list of csv files
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f)
      
    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
      
    # print the content
    print('Content:')
    display(df)
    print()
```

```python
# assign dataset names
list_of_names = ['365_course_info','365_course_ratings','365_exam_info','365_quiz_info','365_student_engagement',
'365_student_exams','365_student_hub_questions','365_student_info','365_student_learning','365_student_purchases'
, '365_student_quizzes']
 
# create empty list
dataframes_list = []
 
# append datasets into the list
for i in range(len(list_of_names)):
    temp_df = pd.read_csv("./"+list_of_names[i]+".csv")
    dataframes_list.append(temp_df)
```
